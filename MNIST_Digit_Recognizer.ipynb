```

import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()


x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)

x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255


num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print(f"x_train shape: {x_train.shape}")
print(f"y_train shape: {y_train.shape}")


model = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(784,), name='Input_Layer'),

    keras.layers.Dropout(0.2),

    keras.layers.Dense(64, activation='relu', name='Hidden_Layer_1'),

    keras.layers.Dropout(0.2),

    keras.layers.Dense(32, activation='relu', name='Hidden_Layer_2'),

    keras.layers.Dense(num_classes, activation='softmax', name='Output_Layer')
])

print("\nModel Architecture:")
model.summary()


model.compile(
    optimizer='adam',

    loss='categorical_crossentropy',

    metrics=['accuracy']
)


print("\nStarting Model Training...")
history = model.fit(
    x_train,
    y_train,
    epochs=20,
    batch_size=128,
    verbose=1,
    validation_data=(x_test, y_test)
)
print("Training Finished.")
```

```

print("\nEvaluating Model Performance on Test Data...")
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
print(f"Test Loss: {loss:.4f}")

print("\nMaking predictions on a few test images...")
predictions = model.predict(x_test)
print("*******")
print(predictions[4])
print("******")
prediction_probabilities = predictions[4]
predicted_label = np.argmax(prediction_probabilities)
true_label = np.argmax(y_test[4])

image = x_test[4].reshape(28, 28)

plt.imshow(image, cmap='gray')
plt.title(f"Prediction: {predicted_label}, True Label: {true_label}")
plt.axis('off')
plt.show()
```

```
for i in np.random.choice(x_test.shape[0], 5):
    # The model's prediction is a probability distribution
    prediction_probabilities = predictions[i]
    # We find the class (digit) with the highest probability
    predicted_label = np.argmax(prediction_probabilities)
    # The true label is also one-hot encoded, so we find its index
    true_label = np.argmax(y_test[i])

    # Reshape the flattened image data back to 28x28 to display it
    image = x_test[i].reshape(28, 28)

    plt.imshow(image, cmap='gray')
    plt.title(f"Prediction: {predicted_label}, True Label: {true_label}")
    plt.axis('off')
    plt.show()
```
